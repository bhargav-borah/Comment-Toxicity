## Preprocessing the data

```python
from tensorflow.keras.layers import TextVectorization
```

```python
X = df['comment_text']
y = df[df.columns[2:]].values
```

```python
X
```

```python
MAX_FEATURES = 200000 # number of words in the vocab
```

```python
vectorizer = TextVectorization(max_tokens=MAX_FEATURES,
                              output_sequence_length=1800,
                              output_mode='int')
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer.get_vocabulary()
```

```python
len(vectorizer.get_vocabulary())
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorized_text
```

```python
vectorized_text.shape
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # helps prevent bottlenecks
```

```python
dataset.as_numpy_iterator().next()
```

```python
batch_X, batch_y = dataset.as_numpy_iterator().next()
```

```python
batch_X.shape
```

```python
batch_y.shape
```

```python
train = dataset.take(int(len(dataset) * 0.7))
val = dataset.skip(int(len(dataset) * 0.7)).take(int(len(dataset) * 0.2))
test = dataset.skip(int(len(dataset) * 0.9)).take(int(len(dataset) * 0.2))
```

```python
train_generator = train.as_numpy_iterator()
```

```python
train_generator.next()
```

## Create Sequential Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model = Sequential()

# Create the embedding layer
model.add(Embedding(MAX_FEATURES+1, 32))

# Create the bidirectional LSTM
model.add(Bidirectional(LSTM(32, activation='tanh')))

# Feature extractor fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))

# Final layer
model.add(Dense(6, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```

## Model Evaluation

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    # Unpack the batch
    X_true, y_true = batch
    
    # Make a prediction
    y_hat = model.predict(X_true)
    
    # Flatten the prediction
    y_true = y_true.flatten()
    y_hat = y_hat.flatten()
    
    pre.update_state(y_true, y_hat)
    re.update_state(y_true, y_hat)
    acc.update_state(y_true, y_hat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall: {re.result().numpy()}, Accuracy: {acc.result().numpy()}')
```
